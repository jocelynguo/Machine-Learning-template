```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
# Splitting the dataset into the Training set and Test set
"""from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)"""
```

```python
# Feature Scaling
"""from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)"""
```

```python
# Fitting Regression model to the dataset
# random forest regression is derived by multiple decision trees

from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=300, random_state=0) #estimator is the number of decision trees

regressor.fit(X, y)
```

```python
regressor.predict(6.5)
```

```python
# Visualising the Linear Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title('Truth or Bluff (random forest)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# more smoother visualization of SVR model
# the salary is the mean of the salary interval
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape(len(X_grid), 1)

# visualize the polynomial regression model
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
plt.title('truth or bluss(random forest)')
plt.xlabel('position')
plt.ylabel('salary')
plt.show()
```
