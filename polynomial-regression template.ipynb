```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
X = dataset.iloc[:, 1:2].values # since X contains only one column data, need to convert X to matrix-like
y = dataset.iloc[:, 2].values
```

```python
# no need to split test and training set, and no need to feature scale since it based on multiple regression
```

```python
# fit linear regression to dataset

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# fit polynomial regression to the dataset

from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)

# include the constant column into X_poly

lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
# visulize the linear regression model
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.title('truth or bluff')
plt.xlabel('position')
plt.ylabel('salaray')
plt.show()
```

```python
# visualize the polynomial regression model
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color='blue')
plt.title('truth or bluss(polynomial reg)')
plt.xlabel('position')
plt.ylabel('salary')
plt.show()
```

```python
# fit polynomial regression to the dataset

from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=3) # change to degree of 3
X_poly = poly_reg.fit_transform(X)

# include the constant column into X_poly

lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
# visualize the polynomial regression model
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color='blue')
plt.title('truth or bluss(polynomial reg)')
plt.xlabel('position')
plt.ylabel('salary')
plt.show()
```

```python
# fit polynomial regression to the dataset

from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=4) # change to degree of 4
X_poly = poly_reg.fit_transform(X)

# include the constant column into X_poly

lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)

# visualize the polynomial regression model
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color='blue')
plt.title('truth or bluss(polynomial reg)')
plt.xlabel('position')
plt.ylabel('salary')
plt.show()
```

```python
####

X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)

# visualize the polynomial regression model
plt.scatter(X, y, color='red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color='blue')
plt.title('truth or bluss(polynomial reg)')
plt.xlabel('position')
plt.ylabel('salary')
plt.show()
```
